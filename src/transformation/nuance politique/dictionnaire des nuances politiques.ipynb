```python
import pandas as pd
from pathlib import Path
from datetime import datetime

```

```python
# 1. Chemins
PROJECT_ROOT = Path.cwd().resolve()
while not (PROJECT_ROOT / "data").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
    PROJECT_ROOT = PROJECT_ROOT.parent

BRONZE_PATH = PROJECT_ROOT / "data" / "bronze" / "bronze_dictionnaire_des_nuances_politiques.csv"
SILVER_PATH = PROJECT_ROOT / "data" / "silver" / "dictionnaire_des_nuances_politiques_silver.csv"
```

```python

# 2. Lecture du Bronze
df_bronze = pd.read_csv(BRONZE_PATH, sep=";")

# 3. Nettoyage et Formatage
df_silver = df_bronze.copy()

# Renommer les colonnes pour matcher tes standards (snake_case)
df_silver = df_silver.rename(columns={
    "Nuance": "id_nuance",
    "Libellé": "libelle_nuance"
})

```

```python
# Nettoyer les chaines de caractères
df_silver["id_nuance"] = df_silver["id_nuance"].str.strip().str.upper()
df_silver["libelle_nuance"] = df_silver["libelle_nuance"].str.strip()

# Ajouter des métadonnées de transformation
df_silver["transformation_timestamp"] = datetime.now().isoformat()
df_silver["source_dataset"] = "bronze_dictionnaire_des_nuances_politiques.csv"

# Sélectionner et réordonner les colonnes pertinentes
colonnes_finales = ["id_nuance", "libelle_nuance", "transformation_timestamp", "source_dataset"]
df_silver = df_silver[colonnes_finales]

# Dédupliquer au cas où
df_silver = df_silver.drop_duplicates(subset=["id_nuance"])

```

```python
# 4. Sauvegarde Silver
df_silver.to_csv(SILVER_PATH, index=False, sep=";", encoding="utf-8")
print(f"✅ Fichier Silver créé : {SILVER_PATH}")
display(df_silver.head())
```
