```python
from pathlib import Path
import pandas as pd

pd.set_option("display.max_rows", 500)
pd.set_option("display.max_columns", 500)
```

```python
project_root = Path.cwd().resolve()
while not (project_root / "data").exists() and project_root != project_root.parent:
    project_root = project_root.parent

bronze_path = project_root / "data" / "bronze" / "bronze_burvot_t2_rhone_69.csv"
silver_commune_path = project_root / "data" / "silver" / "silver_burvot_t2_rhone_69_commune.csv"
silver_candidate_path = project_root / "data" / "silver" / "silver_burvot_t2_rhone_69_candidate.csv"

silver_commune_path.parent.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(
    bronze_path,
    sep=";",
    dtype=str,
    encoding="utf-8"
)

df.columns = [c.strip() for c in df.columns]
display(df.head())

# Fix broken unnamed columns
rename_map = {
    "Unnamed: 26": "N°Panneau.1",
    "Unnamed: 27": "Sexe.1",
    "Unnamed: 28": "Nom.1",
    "Unnamed: 29": "Prénom.1",
    "Unnamed: 30": "Voix.1",
    "Unnamed: 31": "% Voix/Ins.1",
    "Unnamed: 32": "% Voix/Exp.1",
}
df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})

display(df.head())
```

```python
def clean_string(series: pd.Series) -> pd.Series:
    return series.astype("string").str.strip()

def to_int_nullable(series: pd.Series) -> pd.Series:
    return pd.to_numeric(series, errors="coerce").astype("Int64")

def to_float_nullable(series: pd.Series) -> pd.Series:
    return pd.to_numeric(
        series.astype("string").str.replace(",", ".", regex=False),
        errors="coerce"
    ).astype("Float64")
```

```python
# Clean identifiers / strings
string_cols = [
    "Code du département",
    "Libellé du département",
    "Code de la commune",
    "Libellé de la commune",
    "Etat saisie",
    "Sexe",
    "Nom",
    "Prénom",
    "Sexe.1",
    "Nom.1",
    "Prénom.1",
    "extraction_source_url",
    "ingestion_timestamp",
    "source_file_name",
]

for col in string_cols:
    if col in df.columns:
        df[col] = clean_string(df[col])

# preserve leading zeros
if "Code du département" in df.columns:
    df["Code du département"] = df["Code du département"].str.zfill(2)

if "Code de la commune" in df.columns:
    df["Code de la commune"] = df["Code de la commune"].str.zfill(3)


# Cast commune-level numerics
commune_int_cols = [
    "Inscrits",
    "Abstentions",
    "Votants",
    "Blancs",
    "Nuls",
    "Exprimés",
]

commune_float_cols = [
    "% Abs/Ins",
    "% Vot/Ins",
    "% Blancs/Ins",
    "% Blancs/Vot",
    "% Nuls/Ins",
    "% Nuls/Vot",
    "% Exp/Ins",
    "% Exp/Vot",
]

for col in commune_int_cols:
    if col in df.columns:
        df[col] = to_int_nullable(df[col])

for col in commune_float_cols:
    if col in df.columns:
        df[col] = to_float_nullable(df[col])
```

```python
# Build commune-level silver

commune_cols = [
    "Code du département",
    "Libellé du département",
    "Code de la commune",
    "Libellé de la commune",
    "Etat saisie",
    "Inscrits",
    "Abstentions",
    "% Abs/Ins",
    "Votants",
    "% Vot/Ins",
    "Blancs",
    "% Blancs/Ins",
    "% Blancs/Vot",
    "Nuls",
    "% Nuls/Ins",
    "% Nuls/Vot",
    "Exprimés",
    "% Exp/Ins",
    "% Exp/Vot",
    "extraction_source_url",
    "ingestion_timestamp",
    "source_file_name",
]
commune_cols = [c for c in commune_cols if c in df.columns]
df_commune = df[commune_cols].copy()

df_commune["commune_id"] = (
    df_commune["Code du département"].fillna("") + "_"
    + df_commune["Code de la commune"].fillna("")
)

df_commune = df_commune[["commune_id"] + [c for c in df_commune.columns if c != "commune_id"]]
df_commune = df_commune.drop_duplicates(subset=["commune_id"]).reset_index(drop=True)
```

```python
# Build candidate-level silver

candidate_rows = []
suffixes = ["", ".1"]

for suffix in suffixes:
    panneau_col = f"N°Panneau{suffix}"
    sexe_col = f"Sexe{suffix}"
    nom_col = f"Nom{suffix}"
    prenom_col = f"Prénom{suffix}"
    voix_col = f"Voix{suffix}"
    pct_ins_col = f"% Voix/Ins{suffix}"
    pct_exp_col = f"% Voix/Exp{suffix}"

    needed = [panneau_col, sexe_col, nom_col, prenom_col, voix_col, pct_ins_col, pct_exp_col]
    if not all(col in df.columns for col in needed):
        continue

    temp = pd.DataFrame({
        "commune_id": (
            df["Code du département"].fillna("").astype("string") + "_"
            + df["Code de la commune"].fillna("").astype("string")
        ),
        "candidate_slot": ".0" if suffix == "" else suffix,
        "numero_panneau": to_int_nullable(df[panneau_col]),
        "sexe": clean_string(df[sexe_col]),
        "nom": clean_string(df[nom_col]),
        "prenom": clean_string(df[prenom_col]),
        "voix": to_int_nullable(df[voix_col]),
        "pct_voix_ins": to_float_nullable(df[pct_ins_col]),
        "pct_voix_exp": to_float_nullable(df[pct_exp_col]),
        "extraction_source_url": clean_string(df["extraction_source_url"]),
        "ingestion_timestamp": clean_string(df["ingestion_timestamp"]),
        "source_file_name": clean_string(df["source_file_name"]),
    })

    candidate_rows.append(temp)

```

```python
df_candidate = pd.concat(candidate_rows, ignore_index=True)

df_candidate = df_candidate[
    ~(
        df_candidate["numero_panneau"].isna()
        & df_candidate["nom"].isna()
        & df_candidate["prenom"].isna()
    )
].copy()

df_candidate["candidate_id"] = (
    df_candidate["numero_panneau"].astype("string").fillna("") + "_"
    + df_candidate["nom"].fillna("") + "_"
    + df_candidate["prenom"].fillna("")
)

candidate_cols = [
    "commune_id",
    "candidate_id",
    "candidate_slot",
    "numero_panneau",
    "sexe",
    "nom",
    "prenom",
    "voix",
    "pct_voix_ins",
    "pct_voix_exp",
    "extraction_source_url",
    "ingestion_timestamp",
    "source_file_name",
]
df_candidate = df_candidate[candidate_cols].reset_index(drop=True)
```

```python
df_commune.to_csv(silver_commune_path, sep=";", index=False, encoding="utf-8")
df_candidate.to_csv(silver_candidate_path, sep=";", index=False, encoding="utf-8")

print(f"Wrote commune silver: {silver_commune_path}")
print(f"Wrote candidate silver: {silver_candidate_path}")

print("\nCOMMUNE SHAPE:", df_commune.shape)
print("CANDIDATE SHAPE:", df_candidate.shape)

print("\nCOMMUNE SAMPLE:")
display(df_commune.head())

print("\nCANDIDATE SAMPLE:")
display(df_candidate.head(10))
```
