# Exploration des Donnﺣ۸es : ﺣlections Prﺣ۸sidentielles 2022 ﻗ Tour 1 (Bureaux de vote, France entiﺣ۷re)

Ce notebook explore les rﺣ۸sultats du **1er tour** des ﺣ۸lections prﺣ۸sidentielles 2022 au niveau **bureau de vote** pour la **France entiﺣ۷re**, filtrﺣ۸ sur le dﺣ۸partement du Rhﺣﺑne (69).  
Source : [data.gouv.fr ﻗ resultats-par-niveau-burvot-t1-france-entiere.xlsx](https://static.data.gouv.fr/resources/election-presidentielle-des-10-et-24-avril-2022-resultats-definitifs-du-1er-tour/20220414-152612/resultats-par-niveau-burvot-t1-france-entiere.xlsx)  
Identifiant ressource : `98eb9dab-f328-4dee-ac08-ac17211357a8`  
Format : **XLSX** (31.9 Mo) ﻗ converti en **CSV** (`;` comme sﺣ۸parateur)


```python
import pandas as pd
import numpy as np
import os
import requests
from datetime import datetime

pd.set_option('display.max_rows', 500)
```

```python


# ﻗﻗ 1. PATH CONFIGURATION ﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗ
BASE_DATA_DIR = "../../data"

RAW_DIR    = os.path.join(BASE_DATA_DIR, "raw")
BRONZE_DIR = os.path.join(BASE_DATA_DIR, "bronze")

os.makedirs(RAW_DIR, exist_ok=True)
os.makedirs(BRONZE_DIR, exist_ok=True)

XLSX_URL = (
    "https://static.data.gouv.fr/resources/"
    "election-presidentielle-des-10-et-24-avril-2022-resultats-definitifs-du-1er-tour/"
    "20220414-152612/resultats-par-niveau-burvot-t1-france-entiere.xlsx"
)

path_xlsx_raw     = os.path.join(RAW_DIR, "burvot_t1_france_entiere.xlsx")
path_rhone_bronze = os.path.join(BRONZE_DIR, "bronze_burvot_t1_rhone_69.csv")

```

```python
# ﻗﻗ 2. RAW LAYER: Landing Zone ﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗ
# We keep the source file exactly as it is.
if not os.path.exists(path_xlsx_raw):
    print(f"ﻭ۴ Downloading source to RAW...")
    resp = requests.get(XLSX_URL, timeout=180)
    resp.raise_for_status()
    with open(path_xlsx_raw, "wb") as f:
        f.write(resp.content)
    print(f"ﻗ Saved to RAW: {path_xlsx_raw}")
else:
    print(f"ﻗ Raw file already exists in {RAW_DIR}")
```

```python

# ﻗﻗ 3. BRONZE LAYER: Ingestion & Metadata ﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗ
# Bronze is for "raw-ish" data in a readable format with audit columns.
print("\nﻭ  Processing RAW -> BRONZE...")
df_all = pd.read_excel(path_xlsx_raw, header=0, dtype=str, engine="openpyxl")

# Identify the department column
col_dept = next(c for c in df_all.columns if any(kw in c.lower() for kw in ["dﺣ۸partement", "dept"]))

# Filter for Rhﺣﺑne (69)
df_bronze = df_all[df_all[col_dept].astype(str).str.strip().str.lstrip("0") == "69"].copy()

# --- RECONSTRUCT CANDIDATE COLUMNS (Structural Ingestion) ---
CAND_FIELDS = ["NﺡﺍPanneau", "Sexe", "Nom", "Prﺣ۸nom", "Voix", "% Voix/Ins", "% Voix/Exp"]
cols = list(df_bronze.columns)
start_cand = next((i for i, c in enumerate(cols) if "panneau" in c.lower()), None)

if start_cand is not None:
    new_cols = cols[:start_cand]
    # Calculate candidates based on remaining original columns
    n_cand = (len(cols) - start_cand) // len(CAND_FIELDS)
    for k in range(n_cand):
        suffix = f".{k}" if k > 0 else ""
        for field in CAND_FIELDS:
            new_cols.append(f"{field}{suffix}")
    df_bronze.columns = new_cols

# --- ADD BRONZE METADATA ---
# Essential for the Medallion pattern to track data lineage
df_bronze['extraction_source_url'] = XLSX_URL
df_bronze['ingestion_timestamp']   = datetime.now().isoformat()
df_bronze['source_file_name']     = os.path.basename(path_xlsx_raw)

# ﻗﻗ 4. SAVE TO BRONZE ﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗﻗ
df_bronze.to_csv(path_rhone_bronze, index=False, sep=";", encoding="utf-8")

print(f"ﻗ BRONZE dataset created: {path_rhone_bronze}")
print(f"ﻭ Rows: {len(df_bronze)} | Columns: {len(df_bronze.columns)}")
```
