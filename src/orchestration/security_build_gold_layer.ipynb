```python
from pathlib import Path

import pandas as pd

SILVER_SECURITY_PATH = Path("../../data/silver/securite_data_silver_clean.csv")
GOLD_DIR = Path("../../data/gold/security")

print(f"Chargement du fichier : {SILVER_SECURITY_PATH}")
df_sec = pd.read_csv(SILVER_SECURITY_PATH, sep=";", dtype=str)

display(df_sec.head())
print(df_sec.shape)
```

```python
# Normalisation des cles et des types pour alignement avec le modele Gold existant

df_sec = df_sec.copy()

df_sec["code_departement"] = df_sec["code_departement"].astype(str).str.strip().str.zfill(2)
df_sec["code_commune"] = df_sec["code_commune"].astype(str).str.strip().str.zfill(3)
df_sec["code_insee_commune"] = df_sec["code_insee_commune"].astype(str).str.strip().str.zfill(5)

df_sec["annee"] = pd.to_numeric(df_sec["annee"], errors="coerce").astype("Int64")
df_sec["nombre"] = pd.to_numeric(df_sec["nombre"], errors="coerce")
df_sec["taux_pour_mille"] = pd.to_numeric(df_sec["taux_pour_mille"], errors="coerce")
df_sec["insee_pop"] = pd.to_numeric(df_sec["insee_pop"], errors="coerce").astype("Int64")
df_sec["insee_log"] = pd.to_numeric(df_sec["insee_log"], errors="coerce").astype("Int64")
df_sec["complement_info_nombre"] = pd.to_numeric(df_sec["complement_info_nombre"], errors="coerce")
df_sec["complement_info_taux"] = pd.to_numeric(df_sec["complement_info_taux"], errors="coerce")

df_sec["source_dataset_timestamp"] = pd.to_datetime(df_sec["source_dataset_timestamp"], errors="coerce")
df_sec["date_refresh"] = pd.to_datetime(df_sec["date_refresh"], errors="coerce")

df_sec["indicateur"] = df_sec["indicateur"].astype(str).str.strip()
df_sec["unite_de_compte"] = df_sec["unite_de_compte"].astype(str).str.strip()
df_sec["est_diffuse"] = df_sec["est_diffuse"].astype(str).str.strip()
df_sec["source_dataset"] = df_sec["source_dataset"].astype(str).str.strip()

df_sec["id_commune"] = df_sec["code_departement"] + "-" + df_sec["code_commune"]

display(df_sec.head())
print(df_sec.dtypes)
```

```python
# Construction de la dimension des indicateurs de securite

dim_indicateur_securite = (
    df_sec[["indicateur", "unite_de_compte"]]
    .drop_duplicates()
    .sort_values(["indicateur", "unite_de_compte"])
    .reset_index(drop=True)
)

dim_indicateur_securite["id_indicateur_securite"] = [
    f"SEC_{i:03d}" for i in range(1, len(dim_indicateur_securite) + 1)
]

dim_indicateur_securite = dim_indicateur_securite[[
    "id_indicateur_securite",
    "indicateur",
    "unite_de_compte",
]]

display(dim_indicateur_securite.head())
print(dim_indicateur_securite.shape)
```

```python
# Construction de la table de faits securite au grain commune + annee + indicateur

fact_securite = df_sec.merge(
    dim_indicateur_securite,
    on=["indicateur", "unite_de_compte"],
    how="left"
)

fact_securite = fact_securite[[
    "id_commune",
    "annee",
    "id_indicateur_securite",
    "code_departement",
    "code_commune",
    "code_insee_commune",
    "nombre",
    "taux_pour_mille",
    "est_diffuse",
    "insee_pop",
    "insee_log",
    "complement_info_nombre",
    "complement_info_taux",
    "source_dataset",
    "source_dataset_timestamp",
    "date_refresh",
]]

fact_securite = fact_securite.sort_values([
    "code_departement",
    "code_commune",
    "annee",
    "id_indicateur_securite",
]).reset_index(drop=True)

display(fact_securite.head())
print(fact_securite.shape)
```

```python
# Sauvegarde des tables Gold securite

GOLD_DIR.mkdir(parents=True, exist_ok=True)

dim_output_path = GOLD_DIR / "dim_indicateur_securite.csv"
fact_output_path = GOLD_DIR / "fact_securite.csv"

dim_indicateur_securite.to_csv(dim_output_path, sep=";", index=False)
fact_securite.to_csv(fact_output_path, sep=";", index=False)

print(f"Dimension sauvegardee : {dim_output_path}")
print(f"Fact sauvegardee : {fact_output_path}")
```
